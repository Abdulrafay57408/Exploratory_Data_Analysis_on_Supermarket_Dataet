```python
marketFile['Date'] = pd.to_datetime(marketFile['Date'] , format= '%m/%d/%Y')

marketFile['Date']
```

```python
marketFile['Gender_Numeric'] = marketFile['Gender'].map({'Male' :1 , 'Female':0})
marketFile['Gender_Numeric']
```

```python
marketFile['Customer_Type'] = marketFile['Customer type'].map({'Member' : 1 , 'Normal' : 0})

marketFile['Customer_Type']
```

```python
product_line = pd.get_dummies(marketFile['Product line']).astype(int)
product_line
```

```python
marketFile['Product_Line_Numeric'] = marketFile['Product line'].astype('category').cat.codes
```

```python
salesPerBranch = marketFile.groupby('Branch')['Total'].sum()

salesPerBranch
```

```python
avgsalesPerBranch = marketFile.groupby('Branch')['Total'].mean().reset_index()
avgsalesPerBranch
```

```python
plt.pie(salesPerBranch , labels = salesPerBranch.index , autopct='%1.2f%%')

plt.title("Sales Distribution By Branch")

plt.show()
```

```python
sns.barplot(x= "Payment" , y = "Total" ,hue='Payment', data = salesByPaymentMethod)

plt.title("Payment Methods vs Total Sales\n")
plt.xlabel("Payment Methods")
plt.ylabel("Sales")

plt.show()
```

```python
sns.boxplot(data = salesPerGender , x = 'Gender' , y= 'Total' )

plt.title("Sales Distribution by Customer Gender\n")
plt.ylabel("Sales")

plt.show()
```

# **Summary & Insights:**

From the above EDA which conducted on the Supermarket sales on provided Dataset, we extract valuable insights on sales, Customer behavour and more others. Following are the key points which we found:

### **1. Top Branch**

From the Analysis , We found that amoung all branches, **branch** **C** genereates the most sales and and highe Profit.

### **2. Popular Payment Method:**

By the EDA , we found that the **Cash** emerged as the most popular payment method, contributing the largest share of total sales among all other payment methods.

### **3. Gender Influence on Sales:**

Analysis shows that the average sales amount for **Female** customers is higher than the Male Customers.

### **4. Others:**

After EDA we also founds many other valuable insights like:

*   Certain products lines, such as **Home and lifestyle**, **Sports and travel**, produced slightly higher Sales as compare to others.
*   Also observed that Customer Ratings and Sales have a weak correlation.
